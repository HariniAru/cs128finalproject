```python
import numpy as np
from random import randint
```

```python
maze = np.full((40, 15), '#')
ls = []
seenls = []

ix = 20
jy = 8

#Starting opening 1 as cell (0,0)
maze[ix,jy] = '.'
#seenls.append([0,0])
for i in [[ix-1,jy], [ix+1,jy], [ix,jy-1], [ix,jy+1]]:
    ls.append(i)
    seenls.append(i)
    
    
# #Starting opening 2 as cell (len(maze)-1, len(maze[0])-1)
# maze[len(maze)-1, len(maze[0])-1] = '+'
# #seenls.append([0,0])
# for i in [[len(maze)-2, len(maze[0])-1], [len(maze)-1, len(maze[0])-2]]:
#     ls.append(i)
#     seenls.append(i)



while ls != []: # and len(seenls) < len(maze)*len(maze[0]):
    
    ridx = randint(0, len(ls)-1)
    
    #if ls[ridx] not in seenls:
        
    rh = ls[ridx][0]
    rw = ls[ridx][1]

    height = len(maze)
    width = len(maze[0])

    if (rh == 0 and rw == 0):  # TL Corner
        #pass
        lscell = [[0,1], [1,0]]
        cstr = maze[0][1] + maze[1][0]

    elif (rh == 0 and rw == width - 1):  # TR Corner
        #pass
        lscell = [[0,width-2], [1,width-1]]
        cstr = maze[0][width-2] + maze[1][width-1]

    elif (rh == height - 1 and rw == 0):  # BL Corner
        #pass
        lscell = [[height-2,0],[height-1,1]]
        cstr = maze[height-2][0] + maze[height-1][1]

    elif (rh == height - 1 and rw == width - 1):  # BR Corner
        #pass
        lscell = [[height-2,width-1], [height-1,width-2]]
        cstr = maze[height-2][width-1] + maze[height-1][width-2]

    elif (rh == 0):  # T Edge
        #pass
        lscell = [[0,rw-1], [0,rw+1], [1,rw]]
        cstr = maze[0][rw-1] + maze[0][rw+1] + maze[1][rw]

    elif (rw == 0):  # L Edge
        #pass
        lscell = [[rh-1,0], [rh+1,0], [rh,1]]
        cstr = maze[rh-1][0] + maze[rh+1][0] + maze[rh][1]

    elif (rh == height - 1):  # B Edge
        #pass
        lscell = [[height-1,rw-1], [height-1,rw+1], [height-2,rw]]
        cstr = maze[height-1][rw-1] + maze[height-1][rw+1] + maze[height-2][rw]

    elif (rw == width - 1):  #/ R Edge
        #pass
        lscell = [[rh-1,width-1], [rh+1,width-1], [rh,width-2]]
        cstr = maze[rh-1][width-1] + maze[rh+1][width-1] + maze[rh][width-2]

    else:
        lscell = [[rh-1,rw], [rh+1,rw], [rh,rw-1],[rh,rw+1]]
        cstr = maze[rh-1][rw] + maze[rh+1][rw] + maze[rh][rw-1] + maze[rh][rw+1]


    if cstr.count('.') < 2:
        maze[rh][rw] = '.'
        for i in lscell:
            if i not in seenls:
                ls.append(i)
                seenls.append(i)

    ls.remove(ls[ridx])
#     print(maze)
#     print('---')

# print("#: wall, +: space\n")
maze[ix,jy] = 'X'
print(maze)
```
